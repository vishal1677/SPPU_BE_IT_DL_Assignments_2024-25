```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('ecg.csv',header=None)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df=df.drop(140,axis=1)
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
sc=StandardScaler()
df=sc.fit_transform(df)
df
```

```python
x_train,x_test=train_test_split(df,test_size=0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Input
```

```python
encoder=Sequential([
Input(shape=(x_train.shape[1],)),
Dense(32,activation='relu'),
Dense(16,activation='relu'),
Dense(8,activation='relu'),
])

decoder=Sequential([
Input(shape=(8,)),
Dense(16,activation='relu'),
Dense(32,activation='relu'),
Dense(x_train.shape[1],activation='linear')
])

model=Sequential([
    encoder,
    decoder
])
```

```python
import tensorflow as tf
```

```python
model.compile(optimizer='adam',loss=tf.keras.losses.MeanSquaredError)
```

```python
model.fit(x_train,x_train,validation_data=(x_test,x_test),epochs=20,batch_size=30,shuffle=True)
```

```python
pred=model.predict(x_test)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x_test[0],label='Original')
plt.plot(pred[0],label='Reconstructed')
plt.legend()
plt.show()
```

```python
mse=np.mean(np.power(pred-x_test,2),axis=1)
thresh=np.percentile(mse,95)
```

```python
print(thresh)
```

```python
plt.axhline(thresh,linestyle='--',color='r')
plt.plot(mse,marker='o',linestyle='')
```
